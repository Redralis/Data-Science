# Wat is Data Science?

Data science is het proces van data innemen, analyseren en communiceren. Gegevens worden uit een dataset ingelezen en geanalyseerd met behulp van een aantal krachtige tools. Dit wordt dan omgezet door de data scientist naar bruikbare figuren en diagrammen. Ook het zoeken naar correlaties is kenmerkend voor Data Science. Als laatste is machine learning en AI een groot deel van Data Science; er worden voorspellingen gedaan op basis van gegevens. Als laatste een voorbeeld voor het maken van voorspellingen. Dit wordt gedaan op basis van historische gegevens, en is uitgevoerd in assignment 17. Eerst wordt hier een functie (decision tree) getraint die met enige nauwkeurigheid een penguins 'body mass' kan voorspellen op basis van 'bill length' en 'flipper length'. Dit deel van Data Science verschilt met Business Intelligence, omdat er binnen BI vooral gekeken wordt naar wat er slecht/goed ging in vorige periodes. Hiermee wordt dan bedacht wat er nu gedaan kan worden om beter te worden/dezelfde resultaten te behouden. Bij Data Science worden er voorspellingen van hoe het in de toekomst gaat zijn gedaan op basis van dezelfde gegevens.
