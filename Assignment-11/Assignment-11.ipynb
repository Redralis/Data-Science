# Importing...

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

# Numerical vs Categorical bivariate analysis...

```python
penguins.groupby('species').mean()
```

It looks like the species of penguin has an effect on every aspect of it; some have bigger bills, some have more body mass et cetera.

```python
sns.barplot(y="bill_length_mm", x="species", data=penguins)
```

It seems like the Adelie species of penguin have shorter bills compared to the rest of the species. The rest of the species have about the same bill length on average.

```python
sns.barplot(y="bill_depth_mm", x="species", data=penguins)
```

It seems like the Adelie and the Chinstrap species have about the same bill depth. The Gentoo species has a comparatively shallower bill.

```python
sns.barplot(y="flipper_length_mm", x="species", data=penguins)
```

It looks like there isn't much of a difference between flipper length across the species of penguin, but the Gentoo have slightly longer flippers.

```python
sns.barplot(y="body_mass_g", x="species", data=penguins)
```

The subplot reveals that the Adelie and Chinstrap species have about the same body mass, whilst the Gentoo species has a very high body mass on average.
