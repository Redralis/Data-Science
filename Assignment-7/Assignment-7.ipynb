# Creating the histograms of numerical columns and naming the distribution...

```python
pokemon['HP'].plot(kind='hist', bins = 10)
```

The distribution of the HP stat in all pokemon looks like an Extreme Value Type 1 distribution, because the amount of pokemon with low HP is very small compared to the amount of pokemon with a large amount of HP, while the rest of the distribution is normal.

```python
pokemon['Spe'].plot(kind='hist', bins = 10)
```

The distribution of the Speed stat in all pokemon looks like an Extreme Value Type 1 distribution, because the amount of pokemon with low Speed is very small compared to the amount of pokemon with a large amount of Speed, while the rest of the distribution is normal. This seems to be a pattern within pokemon base stats.
